```python
def step1_python(case_path):
    series_IDs = sitk.ImageSeriesReader.GetGDCMSeriesIDs(case_path)
    if not series_IDs:
        raise ValueError("ERROR: given directory \"" + case_path + "\" does not contain a DICOM series.")

    series_file_names = sitk.ImageSeriesReader.GetGDCMSeriesFileNames(case_path, series_IDs[0])
    series_reader = sitk.ImageSeriesReader()
    series_reader.SetFileNames(series_file_names)

    series_reader.MetaDataDictionaryArrayUpdateOn()
    series_reader.LoadPrivateTagsOn()
    itkimage = series_reader.Execute()
    seriesuid = series_reader.GetGDCMSeriesIDs(case_path)
    seriesuid = seriesuid[0]
    case_pixels = sitk.GetArrayFromImage(itkimage) # z,y,x
    spacing = np.array(list(reversed(itkimage.GetSpacing()))) # x,y,z ->z,y,x
    origin = np.array(list(reversed(itkimage.GetOrigin()))) # x,y,z ->z,y,x
        
    return case_pixels, spacing, origin, seriesuid

case_path = r'C:\Users\TMU_AIMC\Desktop\james\honghu\FalsePositiveReduction\data\1.3.6.1.4.1.14519.5.2.1.6279.6001.100225287222365663678666836860'
case_pixels, spacing, origin, seriesuid = step1_python(case_path)
```
